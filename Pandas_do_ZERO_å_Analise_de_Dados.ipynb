***SERIES***:


> S√©ries √© uma matriz rotulada unidimensional capaz de conter qualquer tipo de dados (inteiros, strings, n√∫meros de ponto flutuante, objetos Python, etc.) Os r√≥tulos dos eixos s√£o chamados coletivamente de √≠ndice.

> O m√©todo b√°sico para criar uma s√©rie √©:









```python
s1 = pd.Series(['a', 'b', 'c', 'd'], index=[1, 2, 3, 4], name='Random')
s2 = pd.Series(['e', 'f', 'g', 'h'], index=[1, 2, 3, 4], name='Random2')
```

***DATAFRAMES***:

> DataFrame √© uma estrutura de dados rotulada bidimensional com colunas de tipos potencialmente diferentes. √â como uma planilha ou tabela SQL, ou um dicion√°rio de objetos Series. Geralmente √© o objeto pandas mais comumente usado. Como o Series, o DataFrame aceita muitos tipos diferentes de entradas:





```python
pd.DataFrame([s1, s2])
```
